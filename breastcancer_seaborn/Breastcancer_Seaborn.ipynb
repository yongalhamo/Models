###  Classification using XGBoost (minimal feature selection)

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb           #gradient boosted decision tree classifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(df,y,test_size = 0.3,random_state = 42)

clf1 = xgb.XGBClassifier(random_State = 42)
clf1 = clf1.fit(x_train, y_train)


```

```python
print('Accuracy is : ', accuracy_score(y_test,clf1.predict(x_test)))
cm = confusion_matrix(y_test,clf1.predict(x_test))
sns.heatmap(cm,annot=True,fmt = 'd')
```

### Univariate Feature Selection and XGBoost

```python
 #selectkbest :weeds out features that are independant of the target class
#thus irrelevant to the classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select_feature = SelectKBest(chi2,k=10).fit(x_train,y_train)
print('Score List: ',select_feature.scores_)
print('Feature List:',x_train.columns)
```

```python
x_train_2 = select_feature.transform(x_train)
x_test_2 = select_feature.transform(x_test)

clf2 = xgb.XGBClassifier().fit(x_train_2,y_train)
print("Accuracy:", accuracy_score(y_test,clf2.predict(x_test_2)))
cm2 = confusion_matrix(y_test,clf2.predict(x_test_2))
sns.heatmap(cm2, annot=True, fmt = 'd')
```

###  Recursive Feature Elimination with Cross-Validation

```python
from sklearn.feature_selection import RFECV

clf3 = xgb.XGBClassifier()
rfecv = RFECV(estimator=clf3,step = 1, cv = 5, scoring = 'accuracy',n_jobs=-1).fit(x_train,
                                                                                   y_train)
print("Optimal number of features: ",rfecv.n_features_)
print('Best Features: ', x_train.columns[rfecv.support_])
```

```python
print('accuracy is : ',accuracy_score(y_test,rfecv.predict(x_test)))
```

```python
num_features = [i for i in range(1,len(rfecv.grid_scores_)+1)]
cv_scores = rfecv.grid_scores_
ax = sns.lineplot(x=num_features, y = cv_scores)
ax.set(xlabel = 'No. of selected features', ylabel = 'CV scores')
```

###  Feature Extraction using Principal Component Analysis

```python
#if you have a lot of features then one way to reduce
#feature space is to use a dimentionality reduction technique called PCA

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)

x_train_norm = (x_train - x_train.mean())/(x_train.max()-x_train.min())
x_test_norm = (x_test-x_test.mean())/x_test.max()-x_test.min()
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(x_train_norm)

plt.figure(1, figsize=(10,9))
sns.lineplot(data = np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No. of Components')
plt.ylabel('Cumulative Expcained Variance')
```
